```python
a=school_data_complete.loc[:,["student_name", "school_name", "reading_score", "math_score","budget","grade","type"]]
a.head()
```

```python
grouped_school=a.groupby("school_name")
students_school = grouped_school["student_name"].count()
budget_school=grouped_school["budget"].sum()/grouped_school["student_name"].count()
budget_student = grouped_school["budget"].sum()/grouped_school["student_name"].count()**2
grouped_math_avg = grouped_school["math_score"].sum()/grouped_school["student_name"].count()
grouped_read_avg=grouped_school["reading_score"].sum()/grouped_school["student_name"].count()
math_pass = a.loc[a["math_score"]>=70,:]
grouped_math_pass =  math_pass.groupby("school_name")
total_math_pass = (grouped_math_pass["math_score"].count()/grouped_school["student_name"].count())*100
read_pass = a.loc[a["reading_score"]>=70,:]
grouped_read_pass =  read_pass.groupby("school_name")
total_read_pass = (grouped_read_pass["reading_score"].count()/grouped_school["student_name"].count())*100
overall_rate=((total_read_pass+total_math_pass)/2)
#I am missing the "TYPE" column!!!
a_dif_index= school_data.set_index(["school_name"])["type"]
#grouped_type = a_dif_index.groupby("school_name")
```

```python
top_schools = pd.DataFrame({"#Students":students_school,
                            "Budget":budget_school,
                            "$/student":budget_student,
                           "Math Avg":grouped_math_avg,
                           "Read Avg":grouped_read_avg,
                           "%Math Pass":total_math_pass,
                           "%Read Pass":total_read_pass,
                            "Type":a_dif_index,
                           "%Overall Rate":overall_rate})
z=top_schools
top_schools.style.format({'%Math Pass': "%{:.2f}"}).format({'%Read Pass': "%{:.2f}"}).format({'%Overall Rate': "%{:.2f}"}).format({'Math Avg': "{:.2f}"}).format({'Read Avg': "{:.2f}"}).format({'Budget': "${:.0f}"}).format({'$/student': "${:.0f}"})
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

```python
math_9 = a.loc[a["grade"] == "9th",:]
grouped_math_9=math_9.groupby("school_name")
grouped_math_avg_9=grouped_math_9["math_score"].sum()/grouped_math_9["student_name"].count()
math_10 = a.loc[a["grade"] == "10th",:]
grouped_math_10=math_10.groupby("school_name")
grouped_math_avg_10=grouped_math_10["math_score"].sum()/grouped_math_10["student_name"].count()
math_11 = a.loc[a["grade"] == "11th",:]
grouped_math_11=math_11.groupby("school_name")
grouped_math_avg_11=grouped_math_11["math_score"].sum()/grouped_math_11["student_name"].count()
math_12 = a.loc[a["grade"] == "12th",:]
grouped_math_12=math_12.groupby("school_name")
grouped_math_avg_12=grouped_math_12["math_score"].sum()/grouped_math_12["student_name"].count()
math_schools_grade = pd.DataFrame({"9th Math Average":grouped_math_avg_9,
                            "10th Math Average":grouped_math_avg_10,
                            "11th Math Average":grouped_math_avg_11,
                           "12th Math Average":grouped_math_avg_12})
math_schools_grade.style.format({'9th Math Average': "{:.1f}"}).format({'10th Math Average': "{:.1f}"}).format({'11th Math Average': "{:.1f}"}).format({'12th Math Average': "{:.1f}"})
```
